```python
# imports
import pandas as pd
```

```python
# read in dataset
df = pd.read_csv("datasets/healthcare-dataset-stroke-data.csv")
df.head()
```

```python
# show data types
df.dtypes
```

```python
# describe data
df.describe(include='all')
```

```python
df.count()
```

```python
df.dropna(inplace=True)
```

```python
df.count()
```

```python
# use get_dummies ot encode columns
df = pd.get_dummies(df, columns=['gender', 
                                 'ever_married', 
                                 'work_type', 
                                 'Residence_type', 
                                 'smoking_status'])
df.head()
```

```python
df.count()
```

```python
(df['gender_Other'] == 1).sum()
```

```python
(df['work_type_children'] == 1).sum()
```

```python
# drop extra/not needed feautres
df.drop(['id', 'gender_Other', 'work_type_Never_worked', 'work_type_children'], axis=1, inplace=True)
df.head()
```

```python
# machine leanring 
# create our features
X = df.drop(['stroke'], axis=1)

# create our target
y = df['stroke']
```

```python
# test/train split
# imports
from collections import Counter
from sklearn.model_selection import train_test_split

# split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
Counter(y_train)
```

```python
# oversample
# oversample needed due to imbalance of stroke/no stroke data points
from imblearn.over_sampling import RandomOverSampler

# implement resample
X_resampled, y_resampled = RandomOverSampler(random_state=42).fit_resample(X_train, y_train)
Counter(y_resampled)
```

```python
# use Random Forest Classifier for training model
# imports
from sklearn.linear_model import LogisticRegression

# fit model
model = LogisticRegression(solver='newton-cg', random_state=42)
model.fit(X_resampled, y_resampled)
```

```python
# predictions
# imports
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay
from imblearn.metrics import classification_report_imbalanced
import matplotlib.pyplot as plt

# make predictions
predictions = model.predict(X_test)

# calcuate the accuracy score.
test_acc_score = accuracy_score(y_test, predictions)
train_acc_score = model.score(X_train, y_train)

# define confusion matrix
cm = confusion_matrix(y_test, predictions, labels=model.classes_)

# create a DataFrame from the confusion matrix.
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)

# results
print('RESULTS\n')
print("CONFUSION MATRIX")
disp.plot()
plt.show()

print("ACCURACY SCORES\n")
print(f'Training Accuracy: {train_acc_score}\n')
print(f'Testing Accuracy: {test_acc_score}\n')

print("CLASSIFICATION REPORT\n")
print(classification_report_imbalanced(y_test, predictions))
```
