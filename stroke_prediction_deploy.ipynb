```python
# import
import pandas as pd
```

```python
# bring in data
df = pd.read_csv('datasets/encoded_data.csv')
df.head()
```

```python
# load model
# import
import pickle

# open and load file
with open('model/model.pkl', 'rb') as model:
    model = pickle.load(model)
```

```python
# create new df for model deployment
ml_input_df = df.drop(['stroke'], axis=1)
print(ml_input_df.shape)
```

```python
# make predictions on entire dataset
# imports
import numpy as np

# execute predictions
predictions = model.predict(ml_input_df)
stroke_predictions = np.array(predictions)
print(stroke_predictions)
```

```python
# make predictions a column in the dataframe
df['stroke_prediction'] = stroke_predictions.tolist()
df.head()
```

```python
# save new dataset with predictions
df.to_csv('datasets/prediction_dataset.csv', encoding='utf-8', index=False)
```
